```python
import matplotlib.pyplot as plt

# Accuracy values
frameworks = ['Dask', 'Ray', 'Spark']
accuracies = [78.0, 85.14, 86.7]

# Create a color palette for a fancier look
colors = ['#FFB6C1', '#98FB98', '#FFD700']

# Create bar chart with enhanced styling
plt.figure(figsize=(10, 6))
bar_width = 0.4
bars = plt.bar(frameworks, accuracies, color=colors, width=bar_width, edgecolor='black')

# Add data labels and adjust their position
for bar, accuracy in zip(bars, accuracies):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{accuracy:.2f}%', ha='center', va='bottom', fontsize=10)

# Title and labels with increased font size
plt.title('Accuracy Comparison', fontsize=16)
plt.xlabel('Frameworks', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust y-axis limit and ticks
plt.ylim(0, 100)
plt.yticks(range(0, 101, 10))

# Add a legend
plt.legend(['Accuracy'], loc='upper left')

# Remove spines (borders) for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Precision, Recall, and F1 score values
data = {
    'Framework': ['Ray', 'Spark', 'Dask'],
    'Precision': [84.99, 87.79, 78.22],
    'Recall': [85.14, 86.70, 78.92],
    'F1 Score': [84.49, 86.97, 77.92]
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Set the 'Framework' column as the index
df.set_index('Framework', inplace=True)

# Set a custom color palette
custom_colors = ['#103f5c', '#58508d', '#bc5090']

# Set a custom style using Seaborn
sns.set_style("whitegrid")

# Create a heatmap with custom colors
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap=custom_colors, fmt='.2f', linewidths=.5)
plt.title('Comparison Matrix: Precision, Recall, and F1 Score', fontsize=16)
plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Frameworks', fontsize=12)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.tight_layout()
plt.show()
```
